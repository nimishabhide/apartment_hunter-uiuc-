```python
import pandas as pd
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
from sklearn.metrics import auc, accuracy_score, confusion_matrix, mean_squared_error
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold, RandomizedSearchCV, train_test_split
import numpy as np
import xgboost as xgb
```

```python
a=pd.read_csv('new_uiuc - new_uiuc (4).csv')
b=pd.DataFrame(a)
b
```

```python
b.drop(b.columns[[0,1,2,3,4,28,29,30,31,32,33,34,35]], axis = 1, inplace = True)
b
```

```python
b['Website']= label_encoder.fit_transform(b['Website'])
b['Website'].unique()
b['Posted']= label_encoder.fit_transform(b['Posted'])
b['Posted'].unique()
b['Area']= label_encoder.fit_transform(b['Area'])
b['Area'].unique()
b['Minimum_Beds']= label_encoder.fit_transform(b['Minimum_Beds'])
b['Minimum_Beds'].unique()
b['Maximum_Beds']= label_encoder.fit_transform(b['Maximum_Beds'])
b['Maximum_Beds'].unique()
```

```python
X = b.iloc[:, :22].values
y = b.iloc[:, -1].values
X
```
